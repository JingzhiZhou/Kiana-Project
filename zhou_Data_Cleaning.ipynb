```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
colnames=['Building', 'Level', 'MAC', 'LAT','LNG','Time']
```

```
rio0 = pd.read_csv('/content/gdrive/MyDrive/3 x Zones Group Assignment/Data_ISE534_Spring2021/unzipped/rio_bq_2019000000000000', delimiter = ",",names=colnames, header=None)
rio0 = rio0.replace('"', '', regex=True)
rio0['Building'] = rio0['Building'].str.replace('{Building:', '')
rio0['Level'] = rio0['Level'].str.replace('Level:', '')
rio0['MAC'] = rio0['MAC'].str.replace('ClientMacAddr:', '')
rio0['LAT'] = rio0['LAT'].str.replace('lat:', '')
rio0['LNG'] = rio0['LNG'].str.replace('lng:', '')
rio0['Time'] = rio0['Time'].str.replace('localtime:', '')
rio0['Time'] = rio0['Time'].str.replace(' UTC}', '')
rio0[['Date','Time']] = rio0.Time.str.split(expand=True) 
rio0
```

```
import cv2
```

```
im =  cv2.imread("/content/gdrive/MyDrive/3 x Zones Group Assignment/map/TECA-EXP-1.png")
```

```
im = cv2.resize(im,(1000,500))
imgheight=im.shape[0]
imgwidth=im.shape[1]
```

```
y1 = 0
M = imgheight//10
N = imgwidth//10

for y in range(0,imgheight,M):
    for x in range(0, imgwidth, N):
        y1 = y + M
        x1 = x + N
        tiles = im[y:y+M,x:x+N]

        cv2.rectangle(im, (x, y), (x1, y1), (0, 255, 0))
        cv2.imwrite("save/" + str(x) + '_' + str(y)+".png",tiles)

cv2.imwrite("/content/gdrive/MyDrive/3 x Zones Group Assignment/map/plss.png",im)
```
