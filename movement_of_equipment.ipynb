```python

def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_
```

```python
df=data_sep.groupby(['Building','Level','MAC'])[['LNG']].agg(lambda g: np.percentile(g, 5))  ###5%  95%
df2=data_sep.groupby(['Building','Level','MAC'])[['LNG']].agg(lambda g: np.percentile(g, 95)) 
df3=data_sep.groupby(['Building','Level','MAC'])[['LAT']].agg(lambda g: np.percentile(g, 5))
df4=data_sep.groupby(['Building','Level','MAC'])[['LAT']].agg(lambda g: np.percentile(g, 95))  
```
