```
Aug = pd.read_csv("/content/gdrive/MyDrive/3 x Zones Group Assignment/csv data/rio_aug.csv")
Sep = pd.read_csv("/content/gdrive/MyDrive/3 x Zones Group Assignment/csv data/rio_sep.csv")
Oct = pd.read_csv("/content/gdrive/MyDrive/3 x Zones Group Assignment/csv data/rio_oct.csv")
Nov = pd.read_csv("/content/gdrive/MyDrive/3 x Zones Group Assignment/csv data/rio_nov.csv")
Dec = pd.read_csv("/content/gdrive/MyDrive/3 x Zones Group Assignment/csv data/rio_dec.csv")
```

```
a = Aug.groupby(['MAC','Date']).size().to_frame('counts').reset_index()
b = a.sort_values(by=['counts'], ascending= False, ignore_index=True)
c = pd.value_counts(b['MAC']).to_frame('counts').reset_index()
c
```

```
aa = Sep.groupby(['MAC','Date']).size().to_frame('counts').reset_index()
bb = aa.sort_values(by=['counts'], ascending= False, ignore_index=True)
cc = pd.value_counts(bb['MAC']).to_frame('counts').reset_index()
cc
```

```
a1 = Oct.groupby(['MAC','Date']).size().to_frame('counts').reset_index()
b1 = a1.sort_values(by=['counts'], ascending= False, ignore_index=True)
c1 = pd.value_counts(b1['MAC']).to_frame('counts').reset_index()
c1
```

```
a2 = Nov.groupby(['MAC','Date']).size().to_frame('counts').reset_index()
b2 = a2.sort_values(by=['counts'], ascending= False, ignore_index=True)
c2 = pd.value_counts(b2['MAC']).to_frame('counts').reset_index()
c2
```

```
a3 = Dec.groupby(['MAC','Date']).size().to_frame('counts').reset_index()
b3 = a3.sort_values(by=['counts'], ascending= False, ignore_index=True)
c3 = pd.value_counts(b3['MAC']).to_frame('counts').reset_index()
c3
```

```
equip = pd.merge(c,cc,on='index')
equip = pd.merge(equip,c1,on='index')
equip = pd.merge(equip,c2,on='index')
equip = pd.merge(equip,c3,on='index')
equip.columns = ['Mac','Aug','Sep','Oct','Nov','Dec']
equip
```

```
fa = equip[(equip.Aug >= 10) & (equip.Sep >= 10) & (equip.Oct >= 10) & (equip.Nov >= 10) & (equip.Dec >= 10)]
fa
```
