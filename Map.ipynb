```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Aug = pd.read_csv("/content/gdrive/MyDrive/3 x Zones Group Assignment/Present1_data/rio_aug.csv")
```

```python
equip = pd.read_csv("/content/gdrive/MyDrive/3 x Zones Group Assignment/Present1_data/tps1_level2_equip.csv")
```

```python
a = equip.groupby(['MAC']).size().to_frame('counts').reset_index()
```

```python
b = a['MAC'].to_list()
c = Aug[~Aug.MAC.isin(b)]
c
```

```python
t8 = c.loc[c['Time'].str.startswith('23:00')].copy()
t8[t8['Date']=="2019-08-01"]
```

```python
BBox = (t8.LNG.min(),   t8.LNG.max(),      
         t8.LAT.min(),  t8.LAT.max())
BBox
```

```python
print(type(BBox))
```

```python
ruh_m = plt.imread('/content/gdrive/MyDrive/3 x Zones Group Assignment/map/map1.png')
```

```python
fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(t8.LNG, t8.LAT, zorder=1, alpha= 0.2, c='black', s=10)
ax.set_title('Plotting Data on TPS1 Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```
